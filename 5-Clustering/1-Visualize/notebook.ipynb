# Nigerian Music scraped from Spotify - an analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../data/nigerian-songs.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import seaborn as sns

top = df['artist_top_genre'].value_counts()
plt.figure(figsize=(10,7))
sns.barplot(x=top[:10].index, y = top[:10].values)
plt.xticks(rotation = 45)
plt.title('Top Genre', color = 'blue')
```

```python
df = df[df["artist_top_genre"] != "Missing"]
top = df["artist_top_genre"].value_counts()
sns.barplot(x=top.index, y=top.values)
plt.xticks(rotation=45)
plt.title("Top genres", color="blue")
```

Focusing on Afro Dancehall, Afropop, Nigerian Pop

```python
df = df[
    (df["artist_top_genre"] == "afro dancehall")
    | (df["artist_top_genre"] == "afropop")
    | (df["artist_top_genre"] == "nigerian pop")
]
df = df[df["popularity"] > 0]
top = df["artist_top_genre"].value_counts()
plt.figure(figsize=(10, 7))
sns.barplot(x=top.index, y=top.values)
plt.xticks(rotation=45)
plt.title("Top Genres", color="blue")
```

```python
corrmat = df.corr(numeric_only=True)
f, ax = plt.subplots(figsize = (12, 9))
sns.heatmap(corrmat, vmax=0.8, square=True)
```

```python
sns.set_theme(style="ticks")

g = sns.jointplot(
    data=df,
    x="popularity",
    y="danceability",
    hue="artist_top_genre",
    kind="kde",
)
```

```python
sns.FacetGrid(df, hue="artist_top_genre", height=5) \
    .map(plt.scatter, "popularity", "danceability") \
    .add_legend()
```
